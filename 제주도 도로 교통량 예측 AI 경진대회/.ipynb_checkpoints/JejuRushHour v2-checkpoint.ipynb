```python
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train = df_train.drop(['id'], axis =1)
df_test = df_test.drop(['id'], axis =1)
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

train_encode = df_train.loc[:, ['day_of_week','road_name',
                         'start_node_name', 'start_turn_restricted',
                        'end_node_name', 'end_turn_restricted']]
train_encode.loc[:, :] = train_encode.loc[:, :].apply(LabelEncoder().fit_transform)
df_train_concat = pd.concat([train_encode, df_train.loc[:, ['base_date', 'base_hour', 'lane_count', 'road_rating', 
                                                         'multi_linked', 'connect_code', 'maximum_speed_limit', 'vehicle_restricted',
                                                        'weight_restricted','height_restricted', 'road_type', 'start_latitude',
                                                        'start_longitude', 'end_latitude', 'end_longitude', 'target']]], axis=1)

test_encode = df_test.loc[:, ['day_of_week','road_name',
                         'start_node_name', 'start_turn_restricted',
                        'end_node_name', 'end_turn_restricted']]
test_encode.loc[:, :] = test_encode.loc[:, :].apply(LabelEncoder().fit_transform)
df_test_concat = pd.concat([test_encode, df_test.loc[:, ['base_date', 'base_hour', 'lane_count', 'road_rating', 
                                                         'multi_linked', 'connect_code', 'maximum_speed_limit', 'vehicle_restricted',
                                                        'weight_restricted','height_restricted', 'road_type', 'start_latitude',
                                                        'start_longitude', 'end_latitude', 'end_longitude']]], axis=1)
```

```python
df_test_concat = df_test_concat.astype({'day_of_week' : 'int', 'road_name' : 'int', 'start_node_name' : 'int',
                                       'start_turn_restricted' : 'int', 'end_node_name' : 'int', 'end_turn_restricted' : 'int'})
df_train_concat = df_test_concat.astype({'day_of_week' : 'int', 'road_name' : 'int', 'start_node_name' : 'int',
                                       'start_turn_restricted' : 'int', 'end_node_name' : 'int', 'end_turn_restricted' : 'int'})
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df_train_concat.iloc[:, :-1].values
y = df_train_concat.iloc[:, -1].values
```

```python
count_nan = df_train.isna().sum().sum()
count_nan2 = df_test.isna().sum().sum()

print ('Count of NaN: ' + str(count_nan))
print ('Count of NaN: ' + str(count_nan2))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
                                                   random_state = 0)
```

```python
params = {'learning_rate': 0.005, 
          'max_depth': 20, 
          'boosting_type': 'gbdt', 
          'objective': 'regression', 
          'metric': ['l1','l2'], 
          'is_training_metric': True, 
          'num_leaves': 128, 
          'feature_fraction': 0.9, 
          'bagging_fraction': 0.7, 
          'bagging_freq': 10, 
          'seed':2018,
          'num_iterations': 5000}
```

```python
params = {'learning_rate': 0.005, 
          'max_depth': 20, 
          'boosting_type': 'gbdt', 
          'objective': 'regression', 
          'metric': ['l1','l2'], 
          'is_training_metric': True, 
          'num_leaves': 128, 
          'feature_fraction': 0.9, 
          'bagging_fraction': 0.7, 
          'bagging_freq': 10, 
          'seed':2018,
          'num_iterations': 5000,
          'max_bin' : 255}
```

```python
import lightgbm as lgb

gbm = lgb.LGBMRegressor(**params)
gbm.fit(X_train, y_train,
       eval_set = [(X_test, y_test)],
       eval_metric='l1',
       early_stopping_rounds=1000)
```

```python
import numpy as np
test_pred = np.expm1(gbm.predict(df_test_concat, num_iteration=gbm.best_iteration_))
```

```python
sample_submission['target'] = test_pred
```
